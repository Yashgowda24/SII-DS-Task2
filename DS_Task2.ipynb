### From the given 'Iris' dataset, predict the optimum number of clusters and represent it visually.

#### STEP 1 : Importing all Libraries required

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

#### STEP 2 : Load the iris dataset

```python
iris = datasets.load_iris()
df = pd.DataFrame(iris.data,columns= iris.feature_names)
df.head()
```

#### STEP 3: Finding the optimum number of clusters for k-means classification

```python
x = df.iloc[ :,[0,1,2,3]].values
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(init = "k-means++", n_clusters=i, n_init=10, max_iter=300, random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
#optimum number of clusters would be obtained at the elbow point
#let's plot a line graph of the results obtained and look for the elbow point
plt.plot(range(1,11),wcss)
plt.title("Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

##### At the optimum number of clusters the WCSS does not decrease significantly with each iteration.
##### The optimum number of cluster appears at the elbow point
##### Therefore, it is visible that "3" being the elbow point should be equal to the optimum number of clusters for this data

```python
# Applying Kmeans to the dataset/ Crwating Kmeans classifier
kmeans = KMeans(init = "k-means++", n_clusters=3, n_init=10, max_iter=300, random_state=0)
y = kmeans.fit_predict(x)
#y represents the cluster label

#Visualising the clusuters
plt.scatter( x[y==0,0], x[y==0,1], c="red", s=50, label="Iris-setosa")
plt.scatter( x[y==1,0], x[y==1,1], c="blue", s=50, label="Iris-versicolor")
plt.scatter( x[y==2,0], x[y==2,1], c="green", s=50, label="Iris-virginica")

#Plotting the centroids of the clusuters
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[: ,1],c="yellow", s=50, label="Centroids")

#Describing the elements of the graph
plt.legend()
```

#### Predicted the optimum number of clusters and represented it visually for the given 'Iris' dataset.
